```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
        
# setting this as a global variable so that it can be passed as an argument 
# to the pre-order, in-order and post-order traversal methods
root = None 
def createTree():
    queue = []
    newNodeData = input("Enter the root node: ")
    global root 
    root = Node(newNodeData)
    queue.append(root) 

    while len(queue) != 0:
        parent = queue.pop()
        leftChildData = input(f"Enter the left child of {parent.data} (-1 if there is no left child): ")
        if leftChildData != "-1":
            leftNode = Node(leftChildData)
            parent.left = leftNode
            queue.append(leftNode)

        rightChildData = input(f"Enter the right child of {parent.data} (-1 if there is no right child): ")
        if rightChildData != "-1":
            rightNode = Node(rightChildData)
            parent.right = rightNode
            queue.append(rightNode)


def preOrderTraversal(root):
    if root != None:
        print(root.data, end = '-->')
        preOrderTraversal(root.left)
        preOrderTraversal(root.right)
        
def inOrderTraversal(root):
    if root != None:
        inOrderTraversal(root.left)
        print(root.data, end = '-->')
        inOrderTraversal(root.right)

def postOrderTraversal(root):
    if root != None:
        postOrderTraversal(root.left)
        postOrderTraversal(root.right)
        print(root.data, end = '-->')

        
createTree()
print("Pre-order Traversal: ")
preOrderTraversal(root)
print()
print("In-order Traversal: ")
inOrderTraversal(root)
print()
print("Post-order Traversal:")
postOrderTraversal(root)
```
