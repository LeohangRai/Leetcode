```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
    def __str__(self): #Optional
        return str(self.data)
    
    
        
class LinkedList:
    def __init__(self):
        self.head = None
        #self.tail = None (We could also define the tail pointer like this to keep track of the last node)
    
    def is_empty(self):
        return self.head is None #True if the head is None 
    
    def print_list(self):
        #We have to loop through all the nodes in the SLL and print them
        current = self.head
        #While current is not None:
        while current:
            print(current.data)
            current = current.next
            
    #Insert at the end (append)
    def append(self, data):
        new_node = Node(data)  #Initializing a new node with the data given
    
        #we have 2 possible situations
        #1.The SLL is empty, so head.next is None
        if self.is_empty():
            self.head = new_node

        #2.The SLL is not empty, so we have to loop throuh the SLL to reach the end
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        
    #Insert at the beginning (prepend)    
    def prepend(self, data):
        new_node = Node(data)
        
        if self.is_empty():
            self.head = new_node

        else:
            new_node.next = self.head
            self.head = new_node

            
    def insert_after_node(self, node_to_insert_after, data):
        new_node = Node(data) #initializing the node to be inserted
        
        #First we need to check whether the given node_to_insert_after exists in the SLL or not
        #We can do the following things
        
        #1. First check whether the SLL is empty, if yes then the given node_to_insert_after doesn't exist
        #2. Otherwise loop through all the nodes in the SLL and see if the given node_to_insert_after is there or not
        #There is a minus point here, if the same node (i.e. having same data), this fn will only consider the one 
        #that appears first
        
        if self.is_empty():
            print('The SLL is empty!')
        else:
            #Looping through the elements in the SLL and checking if the given node_to_insert_after exists
            current = self.head
            while current:  #while current is not None:
                if current.data != node_to_insert_after:
                    current = current.next
                else:                    
                    print('The given node to insert after exists in the SLL.')
                    new_node.next = current.next
                    current.next = new_node
                    break
            else:
                print('The given node to insert after does not exist')
            
        
            
mySLL = LinkedList()
mySLL.append('A')
mySLL.append('B')
mySLL.append('C')
mySLL.append('D')
mySLL.insert_after_node('B', 'E')
mySLL.print_list()
print('SLL is empty:',mySLL.is_empty())
print('Head of the SLL:', mySLL.head.data)
mySLL.insert_after_node('M', 'E')
```
